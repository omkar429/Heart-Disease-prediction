```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
incode=ColumnTransformer(transformers=[
    ('Ordinal',OrdinalEncoder(categories=[['Moderate','Heavy']]),[6]),
    ('OneHot',OneHotEncoder(sparse_output=False,drop='first'),[1,5,8,9,10,13,14])
],remainder='passthrough')
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
powertrans=ColumnTransformer(transformers=[
    ('PowerTransforme',PowerTransformer(method='box-cox'),[11,12,13,14,17])
],remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
```

```python
pip=Pipeline([
    ('encod',incode),
    ('power',powertrans)
])
```

```python
x_train=pip.fit_transform(x_train)
x_test=pip.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo=LogisticRegression()
```

```python
lo.fit(x_train,y_train)
```

```python
pre=lo.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,pre)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, pre))
```
